#Q1
Using correlation coefficient function that we used in Pandas assignement ðŸ˜ª,
to calculate corr for the follwoing columns


1.   housing_median_age and median_house_value
2.   total_rooms and total_bedrooms
3.   population and median_house_value
4.   median_income and median_house_value

ðŸ›‘ **Select just 100 point from each columns**

Then, in one figure visualize the relationship between the last columns and let the correlation coefficient value for each chart in axis title

When you finished **export** the figure as fig_1.png

note to save figure before plt.show() ðŸ˜¸

```python
# Select just 100 points from each column
selected_data = df.sample(100)

# Define a function to calculate the correlation coefficient
def calculate_correlation(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    numerator = np.sum((x - x_mean) * (y - y_mean))
    denominator = np.sqrt(np.sum((x - x_mean)**2)) * np.sqrt(np.sum((y - y_mean)**2))
    correlation_coefficient = numerator / denominator
    return correlation_coefficient

# Calculate correlation coefficients for specified columns
corr_housing_median_age_median_house_value = calculate_correlation(selected_data['housing_median_age'], selected_data['median_house_value'])
corr_total_rooms_total_bedrooms = calculate_correlation(selected_data['total_rooms'], selected_data['total_bedrooms'])
corr_population_median_house_value = calculate_correlation(selected_data['population'], selected_data['median_house_value'])
corr_median_income_median_house_value = calculate_correlation(selected_data['median_income'], selected_data['median_house_value'])

# Visualize the relationship between the columns and display correlation coefficient in the title
plt.figure(figsize=(12, 8))

# Plot for housing_median_age and median_house_value
plt.subplot(221)
plt.scatter(selected_data['housing_median_age'], selected_data['median_house_value'], color='blue')
plt.title(f"Correlation Coefficient: {corr_housing_median_age_median_house_value:.2f}")
plt.xlabel('Housing Median Age')
plt.ylabel('Median House Value')

# Plot for total_rooms and total_bedrooms
plt.subplot(222)
plt.scatter(selected_data['total_rooms'], selected_data['total_bedrooms'], color='red')
plt.title(f"Correlation Coefficient: {corr_total_rooms_total_bedrooms:.2f}")
plt.xlabel('Total Rooms')
plt.ylabel('Total Bedrooms')

# Plot for population and median_house_value
plt.subplot(223)
plt.scatter(selected_data['population'], selected_data['median_house_value'], color='green')
plt.title(f"Correlation Coefficient: {corr_population_median_house_value:.2f}")
plt.xlabel('Population')
plt.ylabel('Median House Value')

# Plot for median_income and median_house_value
plt.subplot(224)
plt.scatter(selected_data['median_income'], selected_data['median_house_value'], color='purple')
plt.title(f"Correlation Coefficient: {corr_median_income_median_house_value:.2f}")
plt.xlabel('Median Income')
plt.ylabel('Median House Value')

# Save the figure before displaying
plt.tight_layout()
plt.savefig('fig_1.png')

# Display the plot
plt.show()
```

```python
# Select the specified columns
selected_columns = ['housing_median_age', 'total_rooms', 'total_bedrooms', 'households', 'median_income', 'median_house_value']
selected_data = df[selected_columns]

# Compute the correlation matrix
correlation_matrix = selected_data.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Coefficients Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()

# Save the figure before displaying
plt.savefig('fig_2.png')

# Display the plot
plt.show()
```
